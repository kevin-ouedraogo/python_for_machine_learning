```python
#For instance lets assume that we would like to we would like to show to programatically define a list of element
#that satisfy a very complex criteria that cannot be defined directly using the filter oprions 
#a simple example in our case will be displaying the row where the number of proteines is greater than 2 
#and the ratiio between the the rating and vitamins i sgreater than 2
```

```python
#We define a booean list with no values to start with
items_to_display = []
#we start by itterationg through the list of elemetns of the dataframe 
for index,row in df.iterrows():
    #for each element we will take store the value of protein,rating, and vitamins int heir own variables
    protein = row["protein"]
    vitamins = row["vitamins"]
    rating = row["rating"]
    
    #then we calculate the ratio rating/vitamins
    if vitamins > 0:
        ratio = rating/vitamins
    else:
        ratio = 0
        
    #now we check the value of the ratio and number of protein if the conditions are satisfied we display the row by setting the True 
    #or false in the associated booealn list we will later pass 
    if protein > 2 and ratio > 2:
        items_to_display.append(True)
    else:
        items_to_display.append(False)
        
#when the loop itteration is over we display the result of our boulean list 
print(items_to_display)

#and we pass the booealn list in the [] operation to show t he rows where the data satisfy our define conditions 
df[items_to_display]
```
